 # Исследование платёжеспособности заёмщиков

Кредитный отдел банка поручил проанализировать влияние некоторы факторов на платёжеспособность заёмщиков. Отдел предоставил данные о платёжеспособности клиентов.

Описание данных:
* `children` — количество детей в семье
* `days_employed` — общий трудовой стаж в днях
* `dob_years` — возраст клиента в годах
* `education` — уровень образования клиента
* `education_id` — идентификатор уровня образования
* `family_status` — семейное положение
* `family_status_id` — идентификатор семейного положения
* `gender` — пол клиента
* `income_type` — тип занятости
* `debt` — имел ли задолженность по возврату кредитов
* `total_income` — ежемесячный доход
* `purpose` — цель получения кредита

**Цель исследования** - проверить две гипотезы:

1. Семейное положение клиента влияет на факт своевременного погащения кредита.
2. Количество детей у заёмщика влияент на факт возврата кредита в срок.

**Ход исследования:**

1. Обзор данных
2. Предобработка данных
3. Проверка гипотез

## Предобрадотка данных

### Пропуски и артефакты

Хотя столбы `days_employed` и `total_income` не имеют непосредственного значения для тестирования поставленных гипотез, всё же приведём данные в этих столбцах в порядок.

Для начала разберёмся со столбцом `total_income`. Поскольку столбец содержит количественные значения, заполнять пропуски в нём следует медианным или средним значением.

Посмотрим расхождение между этими двумя значениями:

```python
avg_income = data['total_income'].mean()
median_income = data['total_income'].median()

round(avg_income), round(median_income)
```

Так, среднее значение превышает медианное более чем на 20000. Проверим, есть ли в таблице "выдающиеся значения" дохода:

```python
data.sort_values(by='total_income', ascending=False)['total_income'].round().head(15)
```

У двух клиентов с самым высоким заработком месячный доход существенно выше даже, чем у их ближайших последователей, что говорит о значительном разбросе значений. В целом, у Топ-15 клиентов по уровню заработной платы месячный доход в 10 раз превышает среднее и медианное значения.

В данном случае более корректно заполнить пропуски медианным значением месячного дохода, поскольку оно лучше характеризует всю выборку, отбрасывая экстремально большие и маленькие значения.

Заполним пропуски в столбце `total_income` медианным значением с учётом категории `income_type`:

```python
income_types = data['income_type'].unique()

for value in income_types:
    data.loc[data['income_type'] == value, 'total_income'] = data.loc[data['income_type'] == value, 'total_income'].fillna(
        data.loc[data['income_type'] == value, 'total_income'].median())   

#Проверим:
data['total_income'].isna().sum()
```

Отлично, теперь в столце `total_income` нет пропусков.

Также необходимо исправить значения в столбце `days_employed`. Посмотрим, сколько значений не соотверствуют действительности. Для этого выделим два критерия:
* Количество дней стажа должно быть неотрицательным числом.
* Даже если человек работал 65 лет по 360 дней, его трудовой стаж будет равен 23400 дней. Следовательно, будем отдельно анализировать все значения, превышающие данный порог.

Составим список с "подозрительными" значениями столбца `days_employed`:

```python
suspicious_values = []

#Напишем функцию для заполнения данного списка:
def find_suspicious_values(value):
    if value < 0 or value > 23400:
        suspicious_values.append(value)

data['days_employed'].apply(find_suspicious_values)     
suspicious_values
```

Как видно, в столбце очень много артефактов. Посчитаем количество каждого типа с помощью цикла:

```python
count_negative_values = 0 #количество отрицательных значений в столбце
count_too_large_values = 0 #количество слишком больших значений в столбце

for value in suspicious_values:
    if value < 0:
        count_negative_values += 1
    else:
        count_too_large_values += 1

count_negative_values, count_too_large_values
```

Таким образом, в столбце есть 15906 отрицательных значений и 3445 положительных значений, которые всё равно не соответствуют действительности (стаж в 900 лет без выходных требует невероятного трудолюбия). Посчитаем долю каждого из типов артефактов во всех значениях столбца (19351):

```python
round(count_negative_values/19351, 3), round(count_too_large_values/19351, 3)
```

Выходит, что столбец `days_employed` целиком состоит из артефактов!
В случае с отрицательными значениями, причиной их возникновения могли послужить технические ошибки. В связи с этим можно допустить, что значения, написанные со знаком минус, должны быть положительными. Запишем их по модулю:

```python
data['days_employed'] = data['days_employed'].abs()

#Так как отрицательные значения трудового стажа фигурировали в первых 10 строках таблицы, проверим их ещё раз:
data.head(10)
```

Отрицательные значения ликвидированы.

Теперь поподробнее изучим, у каких категорий людей встречаются экстремально большие значения трудового стажа:

```python
data.loc[data['days_employed'] > 23400, ['days_employed', 'dob_years', 'gender', 'income_type']].head(15)
```

Так, мы выяснили, что такой огромный стаж работы, встречается в таблице у пенсионеров. Заменим экстремальные значения трудового стажа на максимально возможные реалистичные. Будем считать, что клиенты с запредельно большим трудовым стажем начали работать с 18 лет и работали 360 дней в году.

Напишем функцию, которая заменит артефакты на соответствующие значения. Но сначала проверим корректность значений столбца `dob_years`:

```python
min(data['dob_years']), max(data['dob_years'])
```

Очевидно, что заёщику не может быть ноль лет, поэтому проверим, сколько в таблице заёмщиков моложе 18:

```python
data[data['dob_years'] < 18]['dob_years'].count()
```

Так, у многих клиентов в таблице возраст явно не соответствует действительности.

Проверим, информация о каких категориях клиентов есть в датафрейме:

```python
data['income_type'].sort_values().unique()
```

Найдём средний возраст заёмщиков по двум категориям - люди трудоспособного возраста и пенсионеры.

```python
display(round(data.loc[data['income_type'] == 'пенсионер', 'dob_years'].mean()))
round(data.loc[data['income_type'] != 'пенсионер', 'dob_years'].mean())
```

Заполним "странные" значения возраста средними для каждой категории заёмщиков:

```python
data.loc[(data['dob_years'] < 18) & (data['income_type'] == 'пенсионер'), 'dob_years'] = 59
data.loc[(data['dob_years'] < 18) & (data['income_type'] != 'пенсионер'), 'dob_years'] = 40

#Проверим:
data[data['dob_years'] == 0]['dob_years'].count()
```

Неправдоподобных значений в колонке возраст больше нет, значит, можно писать функцию для замены артефактов в столбце `days_employed`.

```python
def lower_days_employed(row):
    
    if row['days_employed'] is not None:
        if row['days_employed'] > 23400:
            row['days_employed'] = (row['dob_years']-18)*360
            return row['days_employed']
        else:
            return row['days_employed']
    else:
        return row['days_employed']

#Проверим функцию на строке:
row = [1234567890, 30]
columns = ['days_employed', 'dob_years']
row = pd.Series(data=row, index=columns)
lower_days_employed(row)

display(row)

#И на датафрейме:
rows = {'days_employed':[1234567890, 5678],
    'dob_years':[67, 40]}
rows = pd.DataFrame(data=rows)
rows['days_employed'] = rows.apply(lower_days_employed, axis=1)

rows
```

```python
#Теперь можно применять функцию к датафрейму:
data['days_employed'] = data.apply(lower_days_employed, axis=1)

data.head(10)
```

Данные выглядят корректно. Но на всякий случай проверим, осталились ли в столбце нереалистичные значения:

```python
data.loc[(data['days_employed']<0)|(data['days_employed']>23400)]['days_employed'].count()
```

Артефактов в столбце больше нет!

Теперь заполним пропуски. Для этого рассчитаем среднее и медианное значения трудового стажа:

```python
avg_days_employed = data['days_employed'].mean()
median_days_employed = data['days_employed'].median()

round(avg_days_employed), round(median_days_employed)
```

Среднее значение трудового стажа более, чем в 2 раза превышает медианное. Учитывая значительный разброс данных, лучше опять же заполнить пропуски медианными значениями. Но заполнять пропуски в столбце будем с учётом возраста клиента, поскольку от него сильно зависит стаж:

```python
data['age_category'] = pd.cut(data['dob_years'], [18, 30, 40, 50, 60, 75])
data['days_employed'] = data['days_employed'].fillna(data.groupby('age_category')['days_employed'].transform('median'))
data = data.drop('age_category', axis=1)

#Проверим:
data['days_employed'].isna().sum()
```

Все пропуски в таблице искоренены!

Посмотрим, какие значения принимает столбец `children`:

```python
data['children'].sort_values().unique()
```

Оказалось, что в данном столбце тоже есть как минимум один артефакт - -1. 20 детей у человека быть, конечно, может, но это единичный случай. Проверим количество таких людей в таблице. 

```python
data[data['children'] == 20]['children'].count()
```

Такая высокая рождаемость уже очень маловероятна. Значит, 20 - это тоже артефакт. Посчитаем количество строк с артифактами в столбце `children`.

```python
data.loc[(data['children'] == -1) | (data['children'] == 5)]['children'].count()
```

Поскольку дополнительных данных о количестве детей у клиентов нет, а число аномальных строк не так велико, чтобы повлиять на исход исследования, удалим их:

```python
data = data.drop(data[(data['children'] == -1) | (data['children'] == 20)].index)

#Проверим:
data['children'].sort_values().unique()
```

Отлично, артфекты полностью удалены!

### Категоризация данных

```python
#Создадим словарь education_dict:
education_dict = data[['education_id', 'education']]
education_dict = education_dict.drop_duplicates().reset_index(drop=True)

education_dict = education_dict.sort_values(by='education_id')
education_dict
```

```python
#Создадим словарь family_dict:
family_dict = data[['family_status_id', 'family_status']]
family_dict = family_dict.drop_duplicates().reset_index(drop=True)

family_dict = family_dict.sort_values(by='family_status_id')
family_dict
```

Удалим столбцы `education` и `family_status` из изначального датафрейма, чтобы он выглядел менее громоздким:

Теперь, когда таблица стала посвободнее, с помощью функции создадим новый столбец `purpose_category` для обобщения целей получения кредита в несколько вышеупомянутых категорий.

```python
def group_purposes(purpose):
    
    if purpose.find('автомобил') != -1:
        return 'операции с автомобилем'
    elif purpose.find('недвижимост') != -1 or purpose.find('жиль') != -1:
        return 'операции с недвижимостью'
    elif purpose.find('свадьб') != -1:
        return 'проведение свадьбы'
    elif purpose.find('образовани') != -1:
        return 'получение образования'

#Проверим функцию:
group_purposes('ремонт автомобиля')
```

```python
#Применим функцию к датафрейму:
data['purpose_category'] = data['purpose'].apply(group_purposes)

#Посмотрим на результат:
data.head(10)
```

Для дальнейшего анализа также сгруппируем клиентов в несколько категорий по уровню дохода. Для этого напишем функцию:

```python
def group_by_income(income):
    
    if income <= 30000:
        return 'E'
    elif 30000 < income <= 50000:
        return 'D'
    elif 50000 < income <= 200000:
        return 'C'
    elif 200000 < income <= 1000000:
        return 'B'
    else:
        return 'A'

#Проверка:
group_by_income(150000)
```

```python
#Создадим новый столбец с категориями дохода клиентов:
income_category = [] #список для категорий дахода

income_category = data['total_income'].apply(group_by_income) #заполним список значениями
data.insert(9, 'total_income_category', income_category) #вставим столбец с категориями дохода клиентов рядом со столбцом total_income
data.head()
```

## Проверка гипотез

### Влияние семейного положения

Согласно первой гипотезе, семейный статус влияет на возврат кредита в срок.
Наличие задолженности по кредиту показывает булевая переменная `debt` (1 - если была задолженность).

Посчитаем общее количество клиентов, не погасивших кредит своевременно:

```python
data['debt'].sum(), data['debt'].mean()
```

Количество клиентов, имевших задолженность по возврату кредитов, составляет 1741 человек, или 8,12% от общего числа заёмщиков.

Для проверки данной гипотезы сгруппируем данные по семейному статусу, чтобы посмотреть, сколько клиентов из разных групп по семейному статусу имели задолженность по кредиту.

```python
def make_pivot(category):
    pivot = data.pivot_table(index=category, values='debt', aggfunc=['sum', 'count', 'mean'])
    pivot = pivot.round(4)
    pivot = pivot.rename(
    columns={'sum':'debtors', 'count':'total', 'mean':'percentage'})
    pivot.columns = pivot.columns.droplevel([1])

    return pivot

fam_status_table = make_pivot('family_status_id')
fam_status_table
```

```python
#Объединим получившуюся таблицу со словарём family_dict:
fam_status_dict = family_dict.merge(fam_status_table, on='family_status_id', how='left')

fam_status_dict
```

Получается, что больше всего клиентов с задолженностью по кредиту средиты женатых/замужних людей, а меньше всего - среди вдов/цов и тех, кто в разводе. Тем не менее это также самая многочисленная и самые малочисленные группы клиентов соответственно.

Однако в относительном выражении самая большая доля людей с задолженностью по возврату кредита в группе тех, кто не женат/не замужем (около 10%), а среди женатых/замужних доля должников такая же, как и по всему датафрейму (около 7,5%). Второе место по количеству должников занимает группа клиентов, находящихся в гражданском браке (9,3%). При этом меньше всего должников среди разведённых и вдов/цов (по 6,6-7%).

В данном случае клиентов из всех групп по семейному статусу можно объединить в две агрегированные группы - тех, у кого есть партнёр (первые две группы по семейному статусу), и тех, у кого нет партнёра (три оставшиеся группы). Поскольку, неочевидно, как объяснить различия в относительном числе должников внутри каждой из этих агрегированных групп, можно предположить, что причинно-следственной связи между семейным статусом и своевременным возратом кредита провести нельзя. Здесь могут играть роль скорее косвенные факторы (получение вдовой/цом наследства, разведёнными супругами - алиментов или части имущества и т.д.).

Учитывая это, можно сказать, что семейное положение не оказывает прямого влияния на факт своевременного погашения кредита, т. е. первая гипотеза подтверждена лишь частично.
