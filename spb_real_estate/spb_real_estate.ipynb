## Предобработка данных

### Пропуски

Сначала разберёмся с пропусками. Начнём со столбцов, где пропущенных значений больше всего. Посмотрим распределение данных в столбце `is_apartment`.

```python
data['is_apartment'].value_counts()
```

```python
data['is_apartment'].mean(), data.loc[data['is_apartment']==True, 'last_price'].count()/data['last_price'].count()
```

Получается, что из заполненных данных апартаменты встречаются всего в 1.8% (или 0.2% от всей выборки), т.е. их продают гораздо реже, чем обычные квартиры. Кроме того, можно предположить, что люди, продающие квартиры, а не апартаменты, просто пропускали данную графу. В связи с этим пропуски можно заполнить значением False. Хотя, учитывая, что графы `studio` и `open_plan` заполнены полностью, возможно, имела место техническая ошибка, и данные стёрлись, например, из-за изменения их формата. Тем не менее заполнить остальные данные значением False представляется единственным возможным вариантом.

Количество апартментов в 0.02% от всей выборки выглядит правдоподобно ещё потому, что, например, квартир-студий 0.6%, а квартир с открытой планировкой около 0.3%:

```python
data['studio'].mean(), data['open_plan'].mean()
```

```python
data['is_apartment'] = data['is_apartment'].fillna(False)

data.head()
```

Теперь посмотрим на значения столбца `balcony`:

```python
data['balcony'].value_counts()
```

У большиства квартир на продажу или нет балконов, или их количество не превышает два. Можно предположить, что некоторые люди, выставившие на продажу квартиры без балкона, просто не стали указывать их число. Значит, в данном столбце можно заполнить пропуски нулём:

```python
data['balcony'] = data['balcony'].fillna(0)

#Проверим количество пропусков в заполненном столбце:
data['is_apartment'].isna().sum(), data['balcony'].isna().sum()
```

Отлично, теперь можно изучать пропуски в других столбцах. Такой же логикой можно заполнить пропуски нулями в столбцах `parks_around3000` и `ponds_around3000` (если человек, продающий квартиру не заполнил данные графы, скорее всего, парков и водоемов в радиусе трёх километров от квартиры нет):

```python
data['parks_around3000'] = data['parks_around3000'].fillna(0)
data['ponds_around3000'] = data['ponds_around3000'].fillna(0)

#Проверим количество пропусков в заполненных столбцах:
data['parks_around3000'].isna().sum(), data['ponds_around3000'].isna().sum()
```

С другими столбцами, в которых есть пропущенные значения, такими как `ceiling_height`, `floors_total`, `locality_name`, `days_exposition`, ситуация сложнее: здесь нельзя предположить, какое значение могло бы быть на месте пропуска, поскольку нельзя проследить никакой закономерности. Средними или медианными значениями эти пропуски заполнить тоже нельзя, поскольку здесь нельзя разделить все данные на несколько понятных категорий, по которым можно было бы посчитать медианное или среднее значение. При этом считать медиану или среднее арифметическое по всей выборке для заполнения пропусков не является корректным, поскольку значения данных столбцов зависят от характеристик определённых квартир, домов, районов, а в данном случае это не будет учтено.

Похожая ситуация наблюдается и со столбцами, значения которых отражают расстояние от квартиры до неких ближайших объектов: `airports_nearest`, `cityCenters_nearest`, `parks_nearest`, `ponds_nearest`: если человек не заполнил, на каком расстоянии от квартиры находится тот или иной объект или данные об этом не были получены автоматически (либо были удалены из-за технической ошибки), невозможно угадать, где расположена квартира. Единственное, в данном случае все квартиры в датасете можно разделить на те, у которых есть парк / пруд в радиусе 3 км, и на те, у которых парка / пруда в радиусе 3 км нет. Тогда можно заполнить пропуски в столбцах `parks_nearest` и `ponds_nearest` медианными значениями, исходя из данных категорий с помощью функции:

```python
def fillna_in_data(row, a, b):
    
    if pd.isna(row[a]) and pd.notna(row[b]):
        if row[b] > 0: 
            return data.loc[data[b] > 0, a].median()
        elif row[b] == 0:
            return data.loc[data[b] == 0, a].median()
    else:
        return row[a]

data['parks_nearest'] = data.apply(fillna_in_data, axis=1, a='parks_nearest', b='parks_around3000')
display(data['parks_nearest'].isna().sum())

data['ponds_nearest'] = data.apply(fillna_in_data, axis=1, a='ponds_nearest', b='ponds_around3000')
data['ponds_nearest'].isna().sum()
```

Пропуски в столбце `parks_nearest` успешно заполнены, однако в столбце `ponds_nearest` количество пропущенных значений не изменилось. Попробуем решить эту проблему: проверим распределение значений в столбце `ponds_around3000`:

```python
data['ponds_around3000'].value_counts()
```

Количество квартир, в радиусе трёх километров от которых нет водоёма, совпадает с количеством пропусков в столбце `ponds_nearest`. Похоже, что все пропуски в данном столбце приходятся на значение 0 в `ponds_around3000`, в связи с чем невозможно посчитать медиану и заполнить пропущенные значения. Проверим эту версию:

```python
data.query('ponds_around3000 == 0')['ponds_nearest'].isna().sum()
```

Гипотеза подтвердилась. Получается, что в данном случае невозможно заполнить пропуски в столбце `ponds_nearest`, поскольку известно только, что водоём в данном случае расположен на расстоянии больше 3км от квартиры. Более точной информации нет.

Отталось проанализировать столбцы, описывающие площадь квартиры: `living_area` и `kitchen_area`. Заполнять пропущенные значения в данных столбцах средним или медианным значением некорректо, поскольку, во-первых, опять же невозможно разделить данные на категории (например, по планировке), от которых зависела бы жилая площадь и площадь кухни в квартире, и, во-вторых, такое заполнение столбцов `living_area` и `kitchen_area` может привести к тому, что какое-то из данных значений или их сумма будет превышать общую площадь определённой квартиры. Однако, учитывая, что для всех квартир известна общая площадь, можно определить, какую долю в среднем составляют от общей площади квартиры жилая площадь и площадь кухни, и заполнить пропуски с учётом этих соотношений. Учитывая, что пропуски в столбцах `living_area` и `kitchen_area` составляют небольшую часть, такое заполнение не исказит данные.

```python
display(data['living_area'].sum()/data[data['living_area'].isna() == False]['total_area'].sum())

data['kitchen_area'].sum()/data[data['kitchen_area'].isna() == False]['total_area'].sum()
```

Таким образом, в среднем жилая площадь стоставляет 57,34% от общей площади квартиры, а площадь кухни - 17,4%. Заполним пропуски в столбцах `living_area` и `kitchen_area` в соответствии с этимми соотношениями.

```python
data['living_area'] = data['living_area'].fillna(round(data['total_area']*0.5734, 2))
data['kitchen_area'] = data['kitchen_area'].fillna(round(data['total_area']*0.174, 2))
```

Пересчитаем количество пропущенных значений:

```python
pd.DataFrame(round((data.isna().mean()*100),2), columns=['NaNs, %']).style.format(
    '{:.2f}').background_gradient('coolwarm')
```

Отлично, все пропуски, которые можно было заполнить, заполнены!

### Изменение типов данных

Теперь изменим тип данных в столбцах, где он, соверешенно очевидно, указан неправильно, т. е. в `first_day_exposition` и `is_apartment`:

```python
# изменим тип данных на datetime в столбце со временем: 
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')

#изменим тип данных на bool в столбце с булевыми значениями:
data['is_apartment'] = data['is_apartment'].astype('bool')
```

Кроме того, во многих столбцах можно заменить тип данных float на int. Так, информацию в столбце `last_price` можно округлить дло рублей, отбросив дробную часть. Помимо этого, во многих столбца выбор типа данных float не отправдан, поскольку они могут содержать только целые значения (например, число балконов `balcony`, число парков в радиусе 3 км `parks_around3000` и т.д.). Однако в некоторых из таких столбцов нельзя изменить тип данных, так как в них остались пропущенные значения (`floors_total` и `days_exposition`).

Чтобы автоматизировать изменение типов данных float на int, напишем цикл:

```python
column_list = ['last_price', 'balcony', 'parks_around3000', 'ponds_around3000']

for column in column_list:
    data[column] = data[column].astype('int')
    
data.head()
```

Данные стали выглядеть понятнее.

### Добавление столбцов

Для последующего анализа добавим в таблицу некоторую расчётную информацию о каждой квартире:

```python
data['price_per_sqm'] = round(data['last_price']/data['total_area'], 2)

data[['last_price', 'total_area', 'price_per_sqm']].head()
```

Отлично. Теперь создадим столбцы с днём недели, месяцем и годом публикации.

```python
data['weekday'] = data['first_day_exposition'].dt.weekday
data['month'] = data['first_day_exposition'].dt.month
data['year'] = data['first_day_exposition'].dt.year

data[['first_day_exposition', 'weekday', 'month', 'year']].head()
```

Напишем функцию, чтобы сгруппировать этажи, на которых находятся квартиры на продажу, в три категории:

```python
def group_by_floor(row):
    
        if row['floor'] == 1:
            return 'первый'
        elif pd.notna(row['floors_total']):
            if row['floor'] == row['floors_total']:
                return 'последний'
            elif row['floor'] != row['floors_total']:
                return 'другой'

#Проверка:
trial = pd.DataFrame(data={
    'floor':[1, 17, 5],
    'floors_total':[12, 17, 10]
})
trial['floor_group'] = trial.apply(group_by_floor, axis=1)
trial
```

```python
#Функция работает, применим её к нашему датафрейму:
data['floor_group'] = data.apply(group_by_floor, axis=1)

data[['floors_total', 'floor', 'floor_group']].head()
```

Осталось вычислить отношения жилой площади и площади кухни к общей площади квартир.

```python
data['living_area_share'] = round(data['living_area']/data['total_area'], 4)
data['kitchen_area_share'] = round(data['kitchen_area']/data['total_area'], 4)

data[['total_area', 'living_area', 'kitchen_area', 'living_area_share', 'kitchen_area_share']].head()
```

Такое явное различие в значениях коэффициента, несмотря на достаточно похожие диаграммы рассеивания, может быть вызвано дискретностью и ограниченным значением числа комнат (только целые числа от 0 до 19) в отличие от значения площади (более широкий диапозон, возможны дробные числа).

Что касается взаимосвязи `last_price` и `city_centers_nearest`, здесь наблюдается обратная картина: средняя цена и ценовой диапозон тем больше, чем меньше расстояние от квартиры до центра города. Однако, судя по диаграмме рассеивания, заивисмость гораздо более слабая, чем в случае с общей площадью и количеством комнат. Это показвает и небольшое значение коэффициента корреляции (-0.2).

Поскольку считать корреляцию для категориальных данных не корректно (зависимость от категории этажа, дня недели, месяца или года нелинейна), построим столбчатую диаграмму по медианным значениям:

```python
hist_columns = ['floor', 'weekday', 'month', 'year']

for column in hist_columns:
    display(column)
    impact_assessment.pivot_table(index=column, values='last_price', aggfunc='median').plot(kind='bar', figsize=(5, 5))
    plt.show()
```

Так, стоимость квартиры никак не зависит от дня недели и месяца размещения объявления. Однако медианная цена квартир, выставленных на продажу в 2014 г., примерно на 40% выше медианных цен квартир, продаваемых в последующие годы. Возможно, в 2015 г. цены на недвижимость упали (но это не точно, нужен дополнительный анализ).

Кроме того, из полученных диаграмм видно, что медианная цена на квартиры, расположенные на 29, 30, 32 и 33 почти в два раза выше, чем медианная цена квартир, расположенных на всех остальных этажах. Однако делать вывод о том, что квартиры на последних этажах стоят дороже, поскольку тогда никак невозможно объяснить "проседание" в цене квартир на 31 этаже.

### Квартиры Санкт-Петербурга

Теперь проанализируем рынок квартир Санкт-Петербурга. Для этого сначала выясним, какие квартиры расположены в центральном районе города.

Определим, какая область входит в центральный район Питера. Для этого вынесем все квартиры в городе в отдельную таблицу и создадим новый столбец, обозначающий расстояние от квартиры до центра города в км:

```python
data_spb = data_filtered.query('locality_name == "Санкт-Петербург"').reset_index(drop=True)
data_spb['km_to_center'] = round(data_spb['city_centers_nearest']/1000)

data_spb.head()
```

Теперь посчитаем среднюю цену за квадратный метр для каждого километра, чтобы определить, как цена зависит от удалённости от центра.

```python
data_spb_km_to_center = round(data_spb.pivot_table(index='km_to_center', values='price_per_sqm'))
data_spb_km_to_center.head()
```

Постоим график зависимости цены от удалённости от центра города:

```python
data_spb_km_to_center.plot(style='o-')
```

Так, из данных видно, что по мере удаления от центра цена на квартиры в Питере падает весьма планомерно. Исключение составялют лишь квартиры, расположенные на 27 км от центра.

Посмотрим, что это за квартиры:

```python
data_spb.query('km_to_center == 27.0')
```

Итак, таких квартир всего две, причём они достаточно сильно различаются по площади и цене - одна квартира дорогая, вторая - весьма дешевая. Из-за этого среднее значение и выбивается из общего тренда.

Учитывая, что на 27-ом км от центра расположено всего две квартиры на продажу с очень разными характеристиками, исключим их из анализа, поскольку считать среднее по ним некорректно. Заменим значение на сренюю цену между 26-м и 28-м км:

```python
data_spb_km_to_center.loc[27, 'price_per_sqm'] = data_spb_km_to_center.loc[[26, 28], 'price_per_sqm'].mean()

data_spb_km_to_center.plot(style='o-', figsize=(7, 5))
```

Такое распределение выглядит логичнее. По графику видно, что цены гораздо выше и сильно варьируются для квартир, расположенных в пределах 9 км от центра Питера. Разброс цен на квартиры, расположенные дальше 9 км от центра города, гораздо меньше. Кроме того, на такие квартиры видно поступательное снижение цены по мере удаления от центра.

Так, будем считать, что в центр города входит область в радиусе 9 км.

Проанализируем сегмент квартир в центре Питера.

```python
round(data_spb.query('km_to_center <= 9')[['last_price', 'total_area', 'rooms', 'ceiling_height', 'city_centers_nearest']].describe(), 1)
```

Так, из описания данных следует, что квартиры в центре Санкт-Петербурга весьма разнородны по всем показателям. Среднестатистическая (медианная) квартира в центре Питера стоит 8,5 млн руб., имеет площадь 72 кв. м. с 3 комнатами и высотой пололка в 3 метра и удалена от центра на 5,3 км.
